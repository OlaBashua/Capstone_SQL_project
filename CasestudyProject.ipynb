```python
import csv
import mysql.connector
from datetime import datetime

# Function to convert date format
def convert_date(date_str):
    # Assuming the current date format is 'M/D/YYYY'
    # Adjust the format string based on your CSV file's date format
    date = datetime.strptime(date_str, '%m/%d/%Y')
    # Convert the date to MySQL format 'YYYY-MM-DD'
    return date.strftime('%Y-%m-%d')

# Open the CSV file
with open(r'C:\Users\bashu\OneDrive\Desktop\python files\fact.csv', 'r') as csv_file:
    reader = csv.reader(csv_file)
    
    # Skip the header row
    next(reader)
    
    # Create a MySQL connection
    cnx = mysql.connector.connect(
    host='localhost',
    user='root',
    password='',
    database='case_study1'
)
    
    # Create a cursor object to interact with the database
    cursor = cnx.cursor()

    # Create the Fact table if it doesn't exist
    create_table_sql = """
    CREATE TABLE IF NOT EXISTS Fact (
        Date DATE,
        ProductId INT,
        Profit DECIMAL(10, 2),
        Sales DECIMAL(10, 2),
        Margin DECIMAL(10, 2),
        COGS DECIMAL(10, 2),
        TotalExpenses DECIMAL(10, 2),
        Marketing DECIMAL(10, 2),
        Inventory INT,
        BudgetProfit DECIMAL(10, 2),
        BudgetCOGS DECIMAL(10, 2),
        BudgetMargin DECIMAL(10, 2),
        BudgetSales DECIMAL(10, 2),
        AreaCode INT,
        PRIMARY KEY (ProductId, Date)
    );
    """
    cursor.execute(create_table_sql)
    
    # Generate SQL insert statements
    for row in reader:
        date = convert_date(row[0])  # Assuming date is in the first column
        product_id = row[1]  # Assuming product ID is in the second column
        profit = row[2]  # Assuming profit is in the third column
        sales = row[3]  # Assuming sales is in the fourth column
        margin = row[4]  # Assuming margin is in the fifth column
        cogs = row[5]  # Assuming COGS is in the sixth column
        total_expenses = row[6]  # Assuming total expenses is in the seventh column
        marketing = row[7]  # Assuming marketing is in the eighth column
        inventory = row[8]  # Assuming inventory is in the ninth column
        budget_profit = row[9]  # Assuming budget profit is in the tenth column
        budget_cogs = row[10]  # Assuming budget COGS is in the eleventh column
        budget_margin = row[11]  # Assuming budget margin is in the twelfth column
        budget_sales = row[12]  # Assuming budget sales is in the thirteenth column
        area_code = row[13]  # Assuming area code is in the fourteenth column
        
        # Generate the SQL insert statement and append it to the list
        sql = f"INSERT INTO Fact (Date, ProductId, Profit, Sales, Margin, COGS, TotalExpenses, Marketing, Inventory, BudgetProfit, BudgetCOGS, BudgetMargin, BudgetSales, AreaCode) " \
              f"VALUES ('{date}', {product_id}, {profit}, {sales}, {margin}, {cogs}, {total_expenses}, {marketing}, {inventory}, {budget_profit}, {budget_cogs}, {budget_margin}, {budget_sales}, {area_code});"
        
        # Execute the SQL statement
        cursor.execute(sql)

    # Commit the changes to the database
    cnx.commit()

    # Close the cursor and MySQL connection
    cursor.close()
    cnx.close()
```

```python
# Create a cursor object to interact with the database
cursor = cnx.cursor()

# Execute the SELECT query to fetch the data from the "Facts" table
select_query = "SELECT * FROM Fact"
cursor.execute(select_query)

# Fetch all the rows from the table
rows = cursor.fetchall()

# Print the fetched data
for row in rows:
    print(row)
```
